```python
def plot_loss(loss_lists, name_list, title_list, ylabel_list):
    colors = ['blue', 'green', 'red', 'cyan', 'magenta', 'black']
    fig, axs = plt.subplots(2, 1, figsize=(10, 10))
    x = np.linspace(1, 30, num=30)
    for i in range(len(loss_lists)):
        for j, loss in enumerate(loss_lists[i]):
            axs[i].plot(x, loss, colors[j], label=name_list[j])
        
        axs[i].set_title(title_list[i])
        axs[i].set_xlabel('Training Epochs')
        axs[i].set_ylabel(ylabel_list[i])
        axs[i].legend()
        
    plt.show()
```

```python
train_loss_list = [torch.load('Adabelief_train_loss.pt'),
                   torch.load('Adabound_train_loss.pt'),
                   torch.load('Adam_train_loss.pt'),
                   torch.load('Lamb_train_loss.pt'),
                   torch.load('SGD_train_loss.pt'),
                   torch.load('Yogi_train_loss.pt')]
name_list = ['Adabelief', 'Adabound', 'Adam', 
             'Lamb', 'SGD', 'Yogi']
```

```python
val_loss_list = [torch.load('Adabelief_val_loss.pt'),
                   torch.load('Adabound_val_loss.pt'),
                   torch.load('Adam_val_loss.pt'),
                   torch.load('Lamb_val_loss.pt'),
                   torch.load('SGD_val_loss.pt'),
                   torch.load('Yogi_val_loss.pt')]
name_list = ['Adabelief', 'Adabound', 'Adam', 
             'Lamb', 'SGD', 'Yogi']
loss_lists = [train_loss_list, val_loss_list]
title_list = ['Training loss in Transformer', 'Validation loss in Transformer']
ylabel_list = ['Training loss', 'Validation loss']

plot_loss(loss_lists, name_list, title_list, ylabel_list)
```
